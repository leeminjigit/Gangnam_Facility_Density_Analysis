```python
error_idx = []

for idx, row in tqdm(df.iterrows()):
    try:
        driver.find_element(By.CSS_SELECTOR, '#nx_query').clear()
        driver.find_element(By.CSS_SELECTOR, '#nx_query').send_keys(row['address'])
        driver.find_element(By.CSS_SELECTOR, '#nx_search_form > fieldset > button > i').click()

        time.sleep(1)

        tmp1 = driver.find_element(By.CSS_SELECTOR, '#no-matched-address-list').text
        tmp2 = driver.find_element(By.CSS_SELECTOR, 'div.uOjIX > div > div.EzDG5 > div.hSYpL').text

        df['address_1'][idx] = tmp1
        df['address_2'][idx] = tmp2

    except:
        df['address_1'][idx] = None
        df['address_2'][idx] = None
        error_idx.append(idx)

driver.quit()
df.head()
```
