```python
import pandas as pd

subway = pd.read_csv('CARD_SUBWAY_MONTH_202306.csv', encoding='utf-8')

subway
```

```python
def unshift_columns(df):
    columns = df.columns.tolist()
    unshifted_columns = columns[1:] + [columns[0]]
    df.columns = unshifted_columns
    return df

subway = unshift_columns(subway)

subway = subway[['역명', '승차총승객수', '하차총승객수']]

subway = subway.set_index('역명')

subway
```

```python
subway = subway.groupby('역명')[['승차총승객수', '하차총승객수']].sum().reset_index()
subway
```

```python
subway_name = pd.read_csv('강남구지하철.csv', index_col=0)
subway_name
```

```python
s_name = subway_name['역명'].unique()
```

```python
subway = subway[subway['역명'].isin(s_name)]
subway['총승객수'] = subway['승차총승객수'] + subway['하차총승객수']

subway
```

```python
# 일 평균 승차 및 하차 총승객수

subway['승차총승객수'] = subway['승차총승객수'] / 30
subway['하차총승객수'] = subway['하차총승객수'] / 30
subway['총승객수'] = subway['총승객수'] / 30


subway['승차총승객수'] = subway['승차총승객수'].round().astype(int)
subway['하차총승객수'] = subway['하차총승객수'].round().astype(int)
subway['총승객수'] = subway['총승객수'].round().astype(int)

subway
```

```python
# 일 평균 승차 및 하차 총승객수

subway['승차총승객수'] = subway['승차총승객수'].round().astype(int)
subway['하차총승객수'] = subway['하차총승객수'].round().astype(int)
subway['총승객수'] = subway['총승객수'].round().astype(int)

subway
```

```python
subway_day = pd.merge(subway_name, subway, on='역명')

subway_day
```

```python
level = [2, 3, 3, 3, 2, 2, 1, 0, 1, 0, 0, 0, 1, 1, 2, 2, 2, 2, 1, 2, 1]
```

```python
subway_day['level'] = level

subway_day
```

```python
subway_day.to_csv('day_subway.csv', encoding='utf-8')
```
